## Keanu Reeves Whoa API

The tests here how I approach testing of an API when an item can be retrieved through an API, but additional information is needed.

I then retrieve the list of movies that contains a 'whoa'.  From there, I am able to retrieve a random `whoa` from each movie that occurs before the initial random `whoa`.  Since each `whoa` object contains information about the movie contains it (e.g. the count of `whoas` in the movie), I am able to calculate the `whoa` index.
